```python
import gym
import HandClassificationEnv
env = gym.make('HandClassificationEnv-v1')
```

```python
# run some tests 

# actions
action = env.action_space.sample()
assert action <9 # ensure action is selecting from 13 cards
    
# observations
obs = env.observation_space.sample()
assert len(obs) == 5 * 32 # 32 bits and 5 cards 
for i in range(len(obs)): assert obs[i] < 2 # all values are 1 or 0
    
# initial observations
initial_obs = env._get_obs()
assert len(initial_obs) == 5 * 32 # 32 bits and 5 cards 
for i in range(len(initial_obs)): assert initial_obs[i] < 2 # all values are 1 or 0

# the initial number of cards 
assert len(env.card_ints) == 5
for i in range(len(env.card_strings)): 
    assert type(env.card_strings[i]) == str
    
# stepping through the environment 
def test_step():
    # get an action
    action = env.action_space.sample()
    
    # take a step with it 
    obs, r, done, _ = env.step(action)
    
    # do some checks
    # obs 
    assert len(obs) == 5 * 32 # 32 bits and 13 cards each 
    for i in range(len(obs)): assert obs[i] < 2 # all values are 1 or 0
        
    # r
    assert r>=-1 and r<= 1
    
    # done
    assert done == True
    
    return r

def test_two():
    action = env.action_space.sample()
    
    # take a step with it 
    obs, r, done, _ = env.step(action)
    
    # reset
    env.reset()
    
    # try again
    obs_2, r_2, done_2, _ = env.step(action)
    
    return 0
    

    
def test_random_reward_distribution():
    from collections import Counter
    print("Random evaluation, 10 000 hands")
    gym.make('HandClassificationEnv-v1')
    rank_classes = []
    actions = []
    reward = 0
    for _ in range(10000):
        env.reset()
        rank_classes.append(env.rank_class)
        action = env.action_space.sample()
        actions.append(action)
        obs, r, done, _ = env.step(action)
        reward += r
        
    print(f"Rank classes: {Counter(rank_classes)}")
    print(f"Predictions: {Counter(actions)}")
    print(f"Reward: {reward}")
    print(f"Average reward: {reward/10000}")
    
    
test_random_reward_distribution()

    
        
test_two()
    
# timing tests for 1000 random steps
print("Timing test")
%time s = sum([test_step() for _ in range(1000)])


assert len(env.reset) == 32 * 5
```
